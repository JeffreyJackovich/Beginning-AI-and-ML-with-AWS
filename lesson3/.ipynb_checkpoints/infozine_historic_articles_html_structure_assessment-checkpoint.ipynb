```python
from bs4 import BeautifulSoup 
import requests

url = 'http://www.infozine.com/news/stories/op/storiesView/sid/1338/'
text = requests.get(url, headers={'User-Agent':'Mozilla/5.0'})
```

```python
soup = BeautifulSoup(text.content)
```

```python
soup
```

```python
topics = soup.find('div', {'class': 'menuTopic'})
```

```python
print(topics)
```

```python
rows = topics.findAll('li')
print(rows)
```

```python
for li in rows:
    link = li.find('a').get('href')  # http://www.infozine.com/news/topics/op/topicsView/tid/5/
    category_page_resp = requests.get(link, headers={'User-Agent':'Mozilla/5.0'}) 
    category_page_soup = BeautifulSoup(category_page_resp.content)
    article_rows = category_page_soup.findAll('div', attrs={"class" : "topicHeadline"})
    #print(article_rows)
    for article_info in article_rows:
        article_url = article_info.find('a').get('href')  # /news/stories/op/storiesView/sid/67176/
        #article_title = article_info.find('a').get('title')
        article_response = requests.get(article_url, headers={'User-Agent':'Mozilla/5.0'})
        article_response_soup = BeautifulSoup(article_response.content)
        article_content_object = article_response_soup.find('div', attrs={"class" : "contentArticle"})
        total_article_content = article_content_object.text
        # slice articel content after (- infoZine -) is 12 characters long
		index_of_infoZine_substring = total_article_content.rfind('- infoZine -')
		index_to_start_slice = index_of_infoZine_substring + 12
		article_content = total_article_content[index_to_start_slice:]
        
        article_title_object = article_response_soup.find('div', attrs={"class" : "articletitle"})
        article_title = article_title_object.text
        
                        
```
